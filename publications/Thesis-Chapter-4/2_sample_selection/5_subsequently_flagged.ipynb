# 5 Subsequently Flagged Samples/Populations

The following samples and or population were subsequently flagged when modelling/evaluating the dataset.

## Setup

```python
import pandas as pd
```

```python
# read in prepared dataset from previous step
df = pd.read_pickle("data/interim/4_sample_order.pkl")
wavelength_cols = df.filter(regex="^\d+", axis=1).columns.tolist()
descriptive_cols = [col for col in df.columns if col not in wavelength_cols]
df
```

## Flag Populations

```python

# create flagged column
df.insert(loc=2, column="subsequent_flag_1", value=0)

# remove populations from test sets







# # instrument 18007 from season 2020 onwards 
# # slope and bias
# df.loc[
#     (df["instrument"] == 18007) & (df["season"] >= 2020),
#     "subsequent_flag_1"
# ] = 4


# instrument 16032 
# consistently under predicting on all three models
df.loc[
    (df["instrument"] == 16032),
    "subsequent_flag_1"
] = 5


# strange fruit, possibly refrigerated
# very low DM for ripened fruit 
df.loc[
    # (df["population"].isin([111200, 111300, 111400, 111500, 111600, 111800, 111900, 112000, 112100, 112200, 112300, 112400, 112500, 112600, 112700])),
    (df["population"].isin([112600, 111900, 111500, 111800, 111200])), # only flag pops within this range that did poorly across all three models after outliers were removed
    "subsequent_flag_1"
] = 1


# populations 112800 and 112900
# populations taken from subsequent days on the same farm in NT- models consistently underpredicting DM
# hypothesized samples were removed from the oven early 
df.loc[
    (df["population"].isin([112800, 112900])),
    "subsequent_flag_1"
] = 2


# population 114100 ripened 
# this population has strangely has both green and ripened fruit 
df.loc[
    (df["population"].isin([114100, 114300])) & (df["physio_stage"] == "ripen"),
    "subsequent_flag_1"
] = 3


# population 110900 ripened 
# consistently under predicting on all three models
df.loc[
    (df["population"] == 110900),
    "subsequent_flag_1"
] = 4


# specific populations where only instruments 15014 and 15038 were used
# DMC is consistently over predicted 
# instrument performed well in populations where multiple other instruments were used
# indicates potential error in methodology for specific persons calibration
df.loc[
    df["population"].isin([114500, 115200, 115000, 122300, 115400, 122200, 115100, 114800, 115300, 114700, 114600]),
    "subsequent_flag_1"
] = 6
```
